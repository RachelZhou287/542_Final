```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfb=pd.read_csv("BiCNN_3stg.csv")
dfb
bbest=dfb["test_acc"].max()
```

```python
dfr=pd.read_csv("ResNet_3stg.csv")
dfr
```

#  ResNet

```python
y=dfr["test_acc"]
x1=dfr["train_loss"]
x2=dfr["test_loss"]
epoch=list(range(1, len(y)+1))
best_acc = y.max()
best_epoch = y.idxmax() + 1   # +1 because epochs start at 1
When=dfr[dfr["test_acc"]==bbest].index+1 # When does ResNet achieve BiCNN's best performance
```

```python
ry=dfr["test_acc"]
r1=dfr["train_loss"]
r2=dfr["test_loss"]
epoch=list(range(1, len(y)+1))
best_acc = ry.max()
best_epoch = ry.idxmax() + 1   # +1 because epochs start at 1


plt.figure(figsize=(10,4))

# ---- Accuracy subplot ----
plt.subplot(1, 2, 1)
plt.plot(epoch, ry, linewidth=2, label="test accuracy")
plt.scatter(best_epoch, best_acc, s=80)  # highlight the point
plt.annotate(
    f"{best_acc:.4f} % at epoch {best_epoch}",
    xy=(best_epoch, best_acc),
    xytext=(best_epoch+1, best_acc),
    #arrowprops=dict(arrowstyle="->", lw=1.5)
)
plt.axvline(When,linestyle="--", linewidth=1,color="purple",label="Best Performance for BiCNN")
plt.legend()
plt.grid(True)

# ---- Loss subplot ----
plt.subplot(1, 2, 2)
plt.plot(epoch, r1, color="blue", label="train loss", linewidth=2)
plt.plot(epoch, r2, color="red", label="test loss", linewidth=2)
plt.legend()
plt.grid(True)
plt.axvline(best_epoch, linestyle="--", linewidth=1)

plt.suptitle("Residual Net", fontsize=18)

plt.tight_layout()
plt.show()

print("best ACC:", best_acc, "%")
```

# BiCNN

```python
by = dfb["test_acc"]
b1 = dfb["train_loss"]
b2 = dfb["test_loss"]
epoch = list(range(1, len(y) + 1))

best_acc = by.max()
best_epoch = by.idxmax() + 1   # +1 because epochs start at 1


plt.figure(figsize=(10,4))

# ---- Accuracy subplot ----
plt.subplot(1, 2, 1)
plt.plot(epoch, by, linewidth=2, label="test accuracy")
plt.scatter(best_epoch, best_acc, s=80)  # highlight the point
plt.annotate(
    f"{best_acc:.4f} % at epoch {best_epoch}",
    xy=(best_epoch, best_acc),
    xytext=(best_epoch+1, best_acc),
    #arrowprops=dict(arrowstyle="->", lw=1.5)
)
plt.legend()
plt.grid(True)

# ---- Loss subplot ----
plt.subplot(1, 2, 2)
plt.plot(epoch, b1, color="blue", label="train loss", linewidth=2)
plt.plot(epoch, b2, color="red", label="test loss", linewidth=2)
plt.legend()
plt.grid(True)
plt.axvline(best_epoch, linestyle="--", linewidth=1)

plt.suptitle("Bilinear CNN", fontsize=18)
plt.tight_layout()
plt.show()

print("best ACC:", best_acc, "%")
```

```python

plt.plot(epoch, by, label="BilinearCNN", linewidth=2)
plt.plot(epoch, ry, label="ResNet", linewidth=2)
plt.title("Accuracy Comparision")

plt.xlabel("Epoch")
plt.ylabel("Test Accuracy (%)")
plt.legend()
plt.grid(True)
plt.show()
```
